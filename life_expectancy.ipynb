1. .***Shape of the Distribution***

    Normal Distribution (Bell-shaped curve): If the histogram has a symmetrical bell shape, the data likely follows a normal distribution.

    ***Skewness:***

        ***Right-skewed (Positive skew):*** If the tail on the right is longer, most values are clustered on the left.

        ***Left-skewed (Negative skew):*** If the tail on the left is longer, most values are clustered on the right.

    ***Uniform Distribution:*** If all bars have roughly the same height, the distribution is uniform, meaning the data is spread evenly across the range.

    ***Bimodal or Multimodal:*** If there are two or more peaks, the data may have two or more distinct groups or modes.

2. ***Central Tendency (Location of Data):***

    ***Peak (Mode):*** The highest point of the histogram shows the mode or the most frequent value.

    ***Mean:*** The average value of the dataset can be inferred from the distribution's center (in a normal distribution, the mean is at the center of the peak).

    ***Median:*** The value that divides the data into two equal halves. In symmetric distributions, the mean and median are close to each other.

3. ***Spread of the Data:***

    ***Range:*** The distance between the minimum and maximum values (visible by the range of the histogram).

    ***Variance/Standard Deviation:*** A wider histogram indicates higher variance or standard deviation, meaning the data points are more spread out.

    ***Interquartile Range (IQR):*** In some cases, looking at the spread around the median can give insight into the IQR.

4. ***Outliers***

    Outliers are values that fall far outside the general distribution of the data (either much higher or much lower than the rest of the values). They might appear as bars separated from the rest or a long tail extending from one side.

***IMPLICATIONS***

***Normal Distribution (Bell-shaped):*** If the histogram is approximately normal, this suggests that many statistical methods (like t-tests or ANOVA) can be applied. For machine learning, models that assume normality (e.g., linear regression) might perform well.

        Action: You can apply parametric statistical tests and linear models.

***Skewed Distribution:***

        Right-skewed (Positive skew): This indicates that there might be a long tail of higher values. Data transformation (e.g., logarithmic transformation) can sometimes help normalize the data.

        Left-skewed (Negative skew): Similar to right-skew, but with the tail on the left. You may also consider transformation to address this skew.

        Action: Consider transformations (log, square root) or use models that handle skewed data well, like decision trees or non-parametric methods.

    Uniform Distribution: Indicates that every value has approximately the same frequency.

        Action: For predictive modeling, uniform distributions might be harder to model effectively, and you might need to group data or rethink how to approach the problem.

    Bimodal/Multimodal: The presence of two or more peaks may suggest multiple subgroups or distributions in the data (e.g., two distinct populations).

        Action: Investigate further. You may need to segment the data or use clustering techniques like k-means or Gaussian mixture models to uncover underlying patterns.

2. Central Tendency:

    Peak (Mode): The mode tells you the most frequent value in the data. If the mode is the most important feature, consider focusing on this value for insights or further analysis.

    Mean vs. Median: If the mean and median are close, the data is likely symmetrical. If they differ significantly, the data might be skewed.

        Action: In case of skewness, consider using the median as a central measure for summary statistics instead of the mean.

3. Spread of the Data:

    Variance/Standard Deviation: A wider histogram suggests greater variability in the data. High variance means the data is spread out, which could influence the choice of model (e.g., regularization might be needed in regression models).

        Action: For high variance, consider regularization techniques (e.g., Lasso, Ridge regression) in machine learning to reduce overfitting.

***BOX PLOTS***
***Median (Central Tendency):***

    The line inside the box represents the median (the 50th percentile) of the data.

        Action: This is the central value of your dataset. If the median is skewed towards one side of the box, it indicates potential skewness in the data.

***2. Interquartile Range (IQR):***

    The box represents the interquartile range (IQR), which spans from the 25th percentile (Q1) to the 75th percentile (Q3).

        Action: The IQR shows the middle 50% of the data. A larger IQR indicates more variability or spread in the data. If the box is very small, the data points are tightly clustered around the median.

***3. Whiskers:***

    The whiskers extend from the edges of the box to the minimum and maximum values that are not outliers.

        Action: If the whiskers are very long, it suggests a larger spread in the data. Whiskers that are very short indicate that most of the data is close to the median.

***4. Outliers:***

    Outliers are typically plotted as individual points outside the whiskers. These are values that are significantly higher or lower than the rest of the data.

        Action: Investigate whether outliers are due to data errors or represent rare but valid extreme values. You may want to remove or treat outliers depending on the context, especially if they distort statistical analysis.

***5. Skewness (Asymmetry):***

    If the boxplot shows a longer whisker on one side, it indicates skewness:

        Right-skewed (positive skew): If the right whisker is longer and/or the box is shifted left, it suggests that most data points are lower, with a few larger values.

        Left-skewed (negative skew): If the left whisker is longer, most data points are higher, with a few smaller values.

        Action: If skewness is present, you may need to transform the data (e.g., log transformation) or use models that are robust to skewness.

***6. Range (Minimum to Maximum):***

    The range of the data is represented by the distance between the minimum and maximum values (excluding outliers).

        Action: A wide range can indicate a diverse dataset with significant variability, while a narrow range suggests data is tightly concentrated.

***7. Comparing Multiple Groups (if applicable):***

    When comparing multiple boxplots (for different groups or categories), you’re interested in:

        Difference in medians: If the medians are different across groups, this suggests potential differences between the groups.

        Overlap of IQRs: If the boxes overlap significantly, there may not be a substantial difference between groups.

 #### ***Action: If the medians differ and the IQRs do not overlap much, you may want to explore statistical tests to compare groups (e.g., t-tests, ANOVA, or non-parametric tests like the Kruskal-Wallis test).***

***8. Symmetry:***

    In symmetric distributions, the box and whiskers should be relatively balanced around the median. If the boxplot is asymmetrical, it indicates skewed data.

        Action: If the distribution is skewed, further data transformations may be necessary for normality or you may opt for models that handle skewed data.

```python
#---------------------------------PEARSON CORRELATION ANALYSIS---------------------------------------

#STEP 2
import pandas as pd
from scipy.stats import pearsonr

# Step 1: Select numerical variables
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
num_cols = num_cols.drop('Life expectancy','Year')  # Exclude target variable

# Step 2: Calculate Pearson correlation with Life expectancy
correlation_results = []

for col in num_cols:
    r, p = pearsonr(df['Life expectancy'].dropna(), df[col].dropna())
    significance = 'Significant' if p < 0.05 else 'Not Significant'
    correlation_results.append({
        'Variable': col,
        'Correlation (r)': round(r, 3),
        'P-Value': round(p, 5),
        'Significance': significance
    })

# Step 3: Create a DataFrame
correlation_df = pd.DataFrame(correlation_results)
correlation_df = correlation_df.sort_values(by='Correlation (r)', ascending=False)

# Display the result
correlation_df.reset_index(drop=True, inplace=True)
print(correlation_df)
```

***Why is Regression Analysis Relevant?***

***Regression analysis is a statistical technique that helps us understand the relationship between one dependent variable (in this case, Life Expectancy) and one or more independent variables (predictors). In the context of this dataset, regression analysis allows us to:***

***Quantify the Relationship:*** While correlation shows the strength and direction of the relationship between variables, regression allows us to quantify it. For example, how much does a 1-unit change in GDP or Schooling influence Life Expectancy?

***Understand the Impact of Multiple Variables:*** We have many predictor variables in this dataset. Regression analysis helps us evaluate how each one impacts Life Expectancy when accounting for other variables. It can show which predictors have the most significant effect when all others are held constant.

***Control for Confounding Factors:*** Some variables may be correlated with others. For instance, GDP and Healthcare Expenditure might be correlated, but regression allows us to assess their individual contributions to life expectancy, controlling for the effect of the other.

***Predict Life Expectancy:*** Once a regression model is built, it can be used to predict Life Expectancy based on the values of the predictors. This can be valuable for understanding how changes in key factors might influence life expectancy in different countries.

***Model Assumptions Testing:*** Regression analysis also helps us check whether assumptions (like linearity, homoscedasticity, normality of residuals, and absence of multicollinearity) hold true for our data, ensuring the validity of our findings.

```python
# Corrected list with the proper column names
selected_features = ['Schooling', 'Income composition of resources', 'BMI', 'Diphtheria', 
                     'Polio', 'GDP', 'Alcohol', 'percentage expenditure', 'Total expenditure', 
                     'Hepatitis B', 'Measles', 'infant deaths', 'under-five deaths', 
                     'thinness 5-9 years', 'thinness  1-19 years', 'HIV/AIDS', 'Adult Mortality']

# Now select the features from the DataFrame
X = df[selected_features]
y = df['Life expectancy']
```

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Scale the features for VIF calculation
X_scaled = StandardScaler().fit_transform(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

# Display VIF results
vif_data
```

***VIF Results Summary:***

The Variance Inflation Factor (VIF) values indicate the level of multicollinearity between the predictor variables. ***A VIF value greater than 10 suggests high multicollinearity, and such variables might need to be removed or regularized.***

Here’s a breakdown of the VIF values from our data:
***Features with High Multicollinearity (VIF > 10)***

    Infant deaths (VIF = 171.18): Extremely high multicollinearity. It might need to be removed from the model or further examined for relevance.

    Under-five deaths (VIF = 172.89): Similar to "infant deaths," it shows high multicollinearity and might need to be removed or adjusted.

    Thinness 5-9 years (VIF = 8.86): Although below the 10 threshold, it’s still on the higher side and could potentially introduce multicollinearity.

    Thinness 1-19 years (VIF = 8.76): Similar to "Thinness 5-9 years," it could be considered for removal due to high VIF.

Other Features (VIF < 10)

    Schooling (VIF = 3.31), Income composition of resources (VIF = 3.02), BMI (VIF = 1.72), Diphtheria (VIF = 2.21), Polio (VIF = 1.95), GDP (VIF = 5.18), Alcohol (VIF = 1.57), Percentage expenditure (VIF = 4.99), Total expenditure (VIF = 1.18), Hepatitis B (VIF = 1.39), Measles (VIF = 1.38), HIV/AIDS (VIF = 1.42), and Adult Mortality (VIF = 1.72) are all within acceptable limits (below 10), indicating that multicollinearity is not a significant issue for these variables.

```python
#Ridge Regression(l2)

from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Define features and target variable
X = df[selected_features]  # Use selected features
y = df['Life expectancy']

# Handle missing data by removing rows with missing values
X = X.dropna()
y = y.dropna()

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize Ridge Regression model
ridge_model = Ridge(alpha=1.0)  # 'alpha' is the regularization strength

# Fit the model
ridge_model.fit(X_train, y_train)

# Predictions and performance evaluation
y_pred = ridge_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Ridge Regression Coefficients:", ridge_model.coef_)
```

```python
from sklearn.linear_model import Lasso

# Initialize Lasso Regression model
lasso_model = Lasso(alpha=0.1)  # 'alpha' is the regularization strength

# Fit the model
lasso_model.fit(X_train, y_train)

# Predictions and performance evaluation
y_pred_lasso = lasso_model.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)

print("Mean Squared Error (Lasso):", mse_lasso)
print("Lasso Regression Coefficients:", lasso_model.coef_)
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit Ridge regression
ridge_model = Ridge(alpha=1.0)  # You can tune alpha
ridge_model.fit(X_scaled, y)

# Get coefficients
ridge_coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': ridge_model.coef_
})

# Rank features by absolute coefficient size (for importance)
ridge_coefficients['Absolute Coefficient'] = ridge_coefficients['Coefficient'].abs()
ridge_coefficients = ridge_coefficients.sort_values(by='Absolute Coefficient', ascending=False)

# Display the ranked features
print(ridge_coefficients[['Feature', 'Coefficient', 'Absolute Coefficient']])

# Fit Lasso regression
lasso_model = Lasso(alpha=0.1)  # You can tune alpha
lasso_model.fit(X_scaled, y)

# Get non-zero coefficients from Lasso
lasso_coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lasso_model.coef_
})

# Filter out features with zero coefficients
lasso_non_zero_features = lasso_coefficients[lasso_coefficients['Coefficient'] != 0]

# Display non-zero features from Lasso
print(lasso_non_zero_features)
```

***Does Life Expectancy has positive or negative correlation with eating habits, lifestyle, exercise, smoking, drinking alcohol etc.***

```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, pearsonr, spearmanr


df = df 
target = 'Life expectancy'

# List lifestyle-related features you want to check correlation with
lifestyle_features = ['BMI', 'Alcohol']  # Adjust column names accordingly

results = []

for feature in lifestyle_features:
    # Drop missing values in target and feature
    data = df[[target, feature]].dropna()
    
    # Shapiro-Wilk test for normality
    stat_target, p_target = shapiro(data[target])
    stat_feat, p_feat = shapiro(data[feature])
    
    # Choose correlation test based on normality
    if p_target > 0.05 and p_feat > 0.05:
        # Both normal - Pearson correlation
        corr, p_val = pearsonr(data[target], data[feature])
        method = 'Pearson'
    else:
        # Not normal - Spearman correlation
        corr, p_val = spearmanr(data[target], data[feature])
        method = 'Spearman'
    
    results.append({
        'Feature': feature,
        'Correlation Method': method,
        'Correlation Coefficient': corr,
        'P-value': p_val,
        'Life Expectancy Normality p': p_target,
        f'{feature} Normality p': p_feat
    })

# Convert results to DataFrame and display
results_df = pd.DataFrame(results)
print(results_df)
```

| Alcohol | Spearman           | **0.429**               | **4.00 × 10⁻¹³²**  | 7.99 × 10⁻²⁹                     | 8.68 × 10⁻³⁷             |

- Both BMI and Alcohol consumption show statistically significant positive correlations with life expectancy (p-value < 0.05).
- The correlation coefficients suggest a moderate positive relationship.
- Normality tests (Shapiro-Wilk) indicate non-normal distribution of life expectancy and these features.
- Spearman correlation is appropriate due to violation of normality assumptions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import shapiro, pearsonr, spearmanr

def statistical_analysis(df, column_name):
    """
    Performs Shapiro-Wilk test for normality, Q-Q plot,
    and Pearson/Spearman correlation between Life expectancy and the selected variable.

    Returns a summary DataFrame.
    """

    # Drop null values
    subset = df[[column_name, 'Life expectancy']].dropna()

    # Q-Q plot
    sm.qqplot(subset[column_name], line='s')
    plt.title(f"Q-Q plot of {column_name}")
    plt.show()

    # Shapiro-Wilk test
    shapiro_stat, shapiro_p = shapiro(subset[column_name])

    # Choose correlation test based on normality
    if shapiro_p > 0.05:
        corr_test = 'Pearson correlation'
        corr_stat, corr_p = pearsonr(subset['Life expectancy'], subset[column_name])
    else:
        corr_test = 'Spearman correlation'
        corr_stat, corr_p = spearmanr(subset['Life expectancy'], subset[column_name])

    # Prepare results
    results = pd.DataFrame({
        'Test': ['Shapiro-Wilk', corr_test],
        'Statistic': [round(shapiro_stat,4), round(corr_stat,4)],
        'p_value': [round(shapiro_p,4), round(corr_p,4)]
    })

    return results
```

```python
#What is the impact of schooling on the lifespan of humans?

results= statistical_analysis(df, 'Schooling')
results
```

To understand how **schooling** influences **life expectancy**, we conducted a statistical analysis that included:

- A **normality test** using the **Shapiro-Wilk** method.
- A **correlation test** using **Spearman correlation** (since the data was not normally distributed).

| Test                   | Statistic | p-value |
|------------------------|-----------|---------|

### 🧠 Interpretation

- **Shapiro-Wilk Test**: The p-value of **0.0000** indicates that the **schooling data is not normally distributed** (p < 0.05).
- **Spearman Correlation**: The correlation coefficient of **0.7794** (p < 0.05) reveals a **strong positive monotonic relationship** between schooling and life expectancy.

### 📌 Conclusion

> The analysis shows that as the **level of schooling increases**, **life expectancy tends to increase as well**. This underscores the importance of education as a **key social determinant of health**.

Investments in education could therefore be a powerful strategy to improve public health and extend lifespan across populations.
